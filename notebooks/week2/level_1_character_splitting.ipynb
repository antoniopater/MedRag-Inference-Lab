# Level 1: Character Splitting

**Opis:** Najprostsza metoda — dzielenie tekstu co stałą liczbę znaków.

**Wady:** Przecina słowa i zdania w połowie — mało użyteczne w produkcji.

**Zastosowanie:** Szybkie prototypowanie.

`chunk size` - liczba, znakow ktora ma zawierac jeden kawalek tekstu. Jest to wartosc statyczna - kady kawalek bedzie mial dokladnie te sama dlugosc.

`chunk overlap` - liczba znakow, ktore powtarzaja sie pomiedzy sasiadujacymi kawalkami. 
Dziala to w ten sposob, ze koncowka pierwszego kawalka pokrywa sie z poczatkiem drugiego kawalka.
Jesli ustawisz chunk na 4 to ostatnie 4 znaki pierwszego kawalka, beda indetyczne jak pierwsze 4 znaki drugiego kawalka.

```python
text = "This is the text I would like to split. It is the example text for the character splitting method."
```

```python
chunks = []

chunk_size = 35 
for i in range(0, len(text), chunk_size):
    chunk = text[i:i+chunk_size]
    chunks.append(chunk)

chunks
```

```python
from langchain_text_splitters import CharacterTextSplitter
chunks_2 = []
text_splitter = CharacterTextSplitter(chunk_size=35, chunk_overlap=4, separator=" ")
text_splitter.create_documents([text])
```

Co to jest `text_splitter` - obiekt, któryy przechowuje regulty dotyczace tego, jak dzielic tekst. 
Mamy nastepujace parametry:
- `chunk_size` - jak duzy ma byc kazdy kawalek
- `chunk_overlap` - ile znakow ma sie powtarzac miedzy kawalkami aby zachowac ciaglosc kontekstu 
- `separator` - znak, ktory wyznacza miejsce ciecia

Czym jest `create_documents()` i po co???
Jest to metoda wywolywana na skonfigurowanym wczesniej obiekcie text_splitter. To ona wykonuje wlasciwa prace.

- jako argument przyjmuje liste tekstow, nawet jak sie ma dlugi tekst trzeba go zamknac w liscie [text]
- jej zadaniem jest pociecie dostarczonego tekstu zgodnie z ustalonumi wczesniej regulami

Funkcja `create_documents` nie zwraca zwyklej listy stringow. Zwraca ona liste obiektow dokumnetu 

`Document Objects` sklada sie z:
- page_content: tutaj znajduje sie wlasciwa tresci pocietego fragmentu
- metadata - to dodatkowe pole, ktore moze przechowywac informacj eo zrodle tekstu, dacue modyfikaacji, numerze strony itd...
